```python
#Function to transform P/L values to float 
def pl2str(x):
    if len(re.findall('%',str(x))) > 0: return round(float(re.sub('\$|\)|\,|%','',re.sub('\(','-',str(x))))/100,3)
    else: return float(re.sub('\$|\)|\,|%','',re.sub('\(','-',str(x))))
```
